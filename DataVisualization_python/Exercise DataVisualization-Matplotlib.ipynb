```python
#Data
import numpy as np 
x = np.arange(0,100)
y = x*2
z = x**2
```

```python

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1.จง plot igure object ด้วยการเรียกใช้ตัวแปร fig ด้วย plt.figure()
#2.create a figure object called fig using plt.figure()
#3.จงใ้ช้ add_axes เพื่อเพิ่ม figure canvas ด้วย ลิสต์ที่เท่ากับ [0,0,1,1] และให้ปรับรุง axis ด้วยตัวแปร ax
#4.จง plot(x,y) บน axes และทำการเติม labels และ title ลงในรูปที่ plot

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
#สร้าง figure และ axes พร้อมกับกำหนดระยะของ axes ทั้งสองอัน ax1 และ ax2 ด้วย [0,0,1,1] และ [0.2,0.5,.2,.2]
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
#จากนั้นทำการสร้าง plot(x,y) บนแกนทั้งสองและ แสดง figure

ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # show figure object
```

```python
#plot ด้วย axes สองชุด พร้อมกำหนดระยะของ axes ทั้งสองด้วย [0,0,1,1] และ [0.2,0.5,.4,.4]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes( [0.2,0.5,.4,.4])
```

```python
#จากนั้นใช้ array x,y,x ทำการ plot บนแกนที่สร้างจากข้อที่แล้ว (notice อย่าลิมกำหนด x - limits และ y- limits)
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

```python
#plt.subplots(nrows = 1 , ncols=2) เพื่อทำการ plot canvas ว่าง
# Empty canvas of 1 by 2 subplots
fig , axes = plt.subplots(nrows = 1 , ncols =2 )
```

```python
#จากนั้นให้ทำการ plot (x,y) และ (x,z) บนแกน axes ของ axes ของ canvas ที่ได้สร้างไว้แล้ว และใช้คำสั่ง linewidth and style เพื่อตกแต่งกราฟ
axes[0].plot(x,y,color = 'blue', lw = 3 ,ls ='--' )
axes[1].plot(x,z,color = 'green' ,lw =3 , ls = '-')

fig
```

```python
#ทำการ resize figure
fig , axes = plt.subplots(nrows = 1 , ncols =2 ,figsize =(12,2) )
axes[0].plot(x,y,color = 'blue', lw = 3 ,ls ='--' )
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color = 'green' ,lw =3 , ls = '-')
axes[1].set_xlabel('x')
axes[1].set_ylabel('x')
```
